```python
new_cols_dict={
    "timestamp":"time",
    "cnt":"count",
    "t1":"temp_real_C",
    "t2":"temp_feels_like_C",
    "hum":"humidity_percent",
    "wind_speed":"wind_speed_kph",
    "weather_code": "weather",
    "is_holiday":"is_holiday",
    "is_weekend":"is_weekend",
    "season":"season"
}

# Renaming the cols as per more technically understandable

bikes_df.rename(new_cols_dict, axis=1,inplace=True)
bikes_df.head()
```

```python
bikes_df.humidity_percent=bikes_df.humidity_percent/100
```

```python
#Creating a season dictionary so that we can map the integers 0-3 to the actual written values

season_dict={
    "0.0":"spring",
    "1.0":"summer",
    "2.0":"autumn",
    "3.0":"winter"
}

#Creating a weather dictionary so that we can map the integers 0-3 to the actual written values

weather_dict={
    "1.0":"clear",
    "2.0":"scattered clouds",
    "3.0":"broken clouds",
    "4.0":"cloudy",
    "7.0":"rain",
    "10.0":"rain with thunder storm",
    "26.0":"snowfall"
}

#changing the season & weather type to string and mapping the corrssponding values:

bikes_df.season=bikes_df.season.astype('str')
bikes_df.season=bikes_df.season.map(season_dict)

bikes_df.weather=bikes_df.weather.astype('str')
bikes_df.weather=bikes_df.weather.map(weather_dict)
```
