```python
most_popular_df=df_tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular_df[:10]
```

```python
df_tracks['release_date']=df_tracks['release_date'].apply(lambda x: x + '-01-01' if len(x) == 4 else x)
df_tracks.set_index("release_date", inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index, errors='coerce')
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda a: round(a/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
corr_df=df_tracks.drop(["id","id_artists","name","artists","key","mode","explicit"],axis=1).corr(method='pearson')
corr_df.head()
```

```python
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Corelation heatmap between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)

```

From this we can find the relation between the fields,
like how how energy is related to loundness 
and from the graph we can see its 0.8 that means if a song is more loud, implies it could be more energetic as well.[-1 means asynchronous]

```python
sample_df=df_tracks.sample(int(0.004*(len(df_tracks))))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="Loudness vs Energy Correlation")
```

```python
sns.regplot(data=sample_df,x="acousticness",y="popularity",color="b").set(title="Acousticnessness vs Popularity Correlation")
```

Here we can see a downward trend between acousticness and popularity that means as the acousticness of a song increases its popluarity decreases.

```python
df_genre=pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```

```python
plt.title("Duration of songs in different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y="genre",x="duration_ms",data=df_genre)
plt.xlabel("Duration in millisecs")
plt.ylabel("Genres")
```

We can see that the songs that belon to Classical and World Genres are the longest in length


```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous= df_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(y="genre",x="popularity",data=famous).set(title="Top 5 Genres by popularity")
```
